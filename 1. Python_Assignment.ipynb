<br>Cosine distance between two points (x,y) and (p,q) is defind as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)

 Hint - If you write the formula correctly you'll get the distance between points (6,-7) and (3,-4) = 0.065

```python
import math
# here S is list of tuples and P is a tuple of len=2
def closest_points_to_p(S, P):
  """
  This function returns the top 5 closest points from list S to given point P, based on Cosine Distance
  """
  lst_distance=[]
  for i in range(len(S)):
    cosine_distance= math.acos((S[i][0]*P[0]+S[i][1]*P[1])/(math.sqrt((S[i][0]**2)+(S[i][1]**2))*math.sqrt((P[0]**2)+P[1]**2)))
    lst_distance.append(cosine_distance)
  
  point_dist=dict(zip(S,lst_distance)) #returns a dictionary where key: value = points in S:distance from P
  asc= sorted(point_dist.items(), key= lambda t:t[1])
  print("The top 5 closest points to P{} are". format(P))
  for j in range(0,5):
    print(asc[j][0]," ", asc[j][1])  
  return closest_points_to_p  # its list of tuples

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
points = closest_points_to_p(S, P)
print() #print the returned values
```

```python
#General equation of a line= ax+by+c
#Logic- if the distance of all red points from the line are positive and the distance of all blue points from the line are negative (or vice versa), print YES, else print NO.

import re
import math
def i_am_the_one(red,blue,lines):
  """
  This function prints YES if all red points are present on one side of the line (in lines) and blue points are present on the other side of the line, else it prints NO
  """
  #First we need to parse the coefficients of the line equations
  lst=[]
  for i in range(len(lines)):
    for j in range(len(lines[i])):
      l= [float(j) for j in re.split('[xXyY]', lines[i])] #Reference= https://stackoverflow.com/questions/56948506/how-to-extract-coefficients-from-a-line-equation-in-python-without-using-numpy/61547364#61547364
    lst.append(l)

  for i in lst:
    distance_red=[]
    distance_blue=[]
    for j in red:
      distance_from_line_red= (i[0]*j[0]+ i[1]*j[1]+i[2])/math.sqrt((i[0]**2)+(i[1]**2)) #Formula= Perpendicular distance of a point from a line of the form ax+by+c- https://en.wikipedia.org/wiki/Distance_from_a_point_to_a_line
      distance_red.append(distance_from_line_red)
    for k in blue:
      distance_from_line_blue= (i[0]*k[0]+ i[1]*k[1]+i[2])/math.sqrt((i[0]**2)+(i[1]**2))
      distance_blue.append(distance_from_line_blue)
    if all(a>0 for a in distance_red) and all(b<0 for b in  distance_blue):
      print("YES")
    elif all(a<0 for a in distance_red) and all(b>0 for b in  distance_blue):
      print("YES")
    else:
      print("NO")
  return ("**********************")

Lines=["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]
Red= [(1,1),(2,1),(4,2),(2,4), (-1,4)]
Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
print(i_am_the_one(Red,Blue,Lines))
```

Ex 2: 40, _, _, _, 60 ==> (60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5 ==> 20, 20, 20, 20, 20 i.e. the sum of (60+40) is distributed qually to all 5 places<br>

Ex 4: _, _, 30, _, _, _, 50, _, _  
==> we will fill the missing values from left to right 
    a. first we will distribute the 30 to left two missing values (10, 10, 10, _, _, _, 50, _, _)
    b. now distribute the sum (10+50) missing values in between (10, 10, 12, 12, 12, 12, 12, _, _) 
    c. now we will distribute 12 to right side missing values (10, 10, 12, 12, 12, 12, 4, 4, 4)
</pre>
for a given string with comma seprate values, which will have both missing values numbers like ex: "_, _, x, _, _, _"
you need fill the missing values

Q: your program reads a string like ex: "_, _, x, _, _, _" and returns the filled sequence

Ex: 
<pre>
Input1: "_,_,_,24"
Output1: 6,6,6,6

Input2: "40,_,_,_,60"
Output2: 20,20,20,20,20

Input3: "80,_,_,_,_"
Output3: 16,16,16,16,16

Input4: "_,_,30,_,_,_,50,_,_"
Output4: 10,10,12,12,12,12,4,4,4

</pre>



```python

def curve_smoothing(string):
  """
  This Function returns a smoothed string according to the given format.
  """
  #convert all numbers in the string from str to float 
    lst=string.split(',')
    for i in range(len(lst)):
        if lst[i]!='_':
            lst[i]=float(lst[i])
    #Left Boundary case= ("_,_,_,number)
    for index_1 in range(len(lst)):
        if type(lst[index_1])!=str:
            for a in range(index_1 + 1):
                lst[a]=(lst[index_1])/(index_1 + 1)
            index_value=(lst[index_1])
            next_index=index_1
            break

    #Middle case= (number,_,_,_,number)
    for index_2 in range(next_index + 1,len(lst)):
        if type(lst[index_2])!=str:
            temp_value=(index_value+lst[index_2])/(index_2-next_index+1)
            for b in range(next_index,index_2+1):
                lst[b]=temp_value
            index_value=(lst[index_2])    
            next_index=index_2

    #Right Boundary case= (number,_,_,_,_")
    count=0
    for index_3 in range(next_index + 1,len(lst)):
        if type(lst[index_3])!=float:
            count=lst.count('_')
            break
    value=index_value/(count+1)
    for index_3 in range(next_index,len(lst)):
        lst[index_3]=value
    
    #convert list back to string
    string2 = ' '.join([str(i) for i in lst])
    return string2
    

S= "_,_,30,_,_,_,50,_,_"
print(curve_smoothing(S))
```

<h3> Q10: Given two sentances S1, S2</h3>
You will be given a list of lists, each sublist will be of length 2 i.e. [[x,y],[p,q],[l,m]..[r,s]]
consider its like a martrix of n rows and two columns

a. the first column Y will contain interger values <br>
b. the second column $Y_{score}$ will be having float values <br>
Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.4243099
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python
import math
def compute_log_loss(A):
  """
  This function returns the log loss of a matrix of size n*2
  """
    sum=0
    for i in A:
      f= (i[0]*math.log10(i[1]))+((1-i[0])*math.log10(1-i[1]))
      sum+=f
    loss = (-1)*(1/len(A))*(sum)
    return loss

A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
loss = compute_log_loss(A)
print(loss)
```
