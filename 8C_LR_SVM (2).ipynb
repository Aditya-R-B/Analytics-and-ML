```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X= b * np.sin(phi)
Y= a * np.cos(phi)
```

```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[0]*mi - intercept)/coef[0]), mi],[((-coef[0]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1], label='Hyperplane')
    plt.legend()
```

```python
#Here X and Y are the coordinates of each point on the Ellipse
alphas=[0.0001,1,100] #Regularizer
outliers= [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
plt.figure(figsize=(20,20))
j,count=0,0
for ind,al in enumerate(alphas):
  X_1=X #Assigning new values for the loop
  Y_1=Y # "  
  for ind_1,out in enumerate(outliers):
    count+=1 #To count the number of outliers in each iteration
    j+=1
    X_1=np.append(X_1,out[0]).reshape(-1,1) #Appending the outlier coordinate of X
    Y_1=np.append(Y_1,out[1]).reshape(-1,1) #Appending the outlier coordinate of Y
    clf=SGDRegressor(alpha=al,eta0=0.001, learning_rate='constant',random_state=0)
    clf.fit(X_1,Y_1)
    y_pred=clf.predict(X_1)
    plt.subplot(3, 5, j)
    plt.title("alpha={},Number of outliers={}".format(alphas[ind],count))
    plt.scatter(X_1,Y_1,color='Black')
    plt.plot(X_1,y_pred)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
  count=0
plt.show()
```

### OBSERVATION

Here we are dealing with a Linear Regression problem where we are trying to minimize the 'square loss'.

* **Case-1** when **Alpha**(Regularization Parameter)**= 0.001**: 
Here the Regularization strength is very weak. Hence more importance will be given to the loss term which tries to minimize the loss and overfit the model without any errors. Hence it is trying to fit all the datapoints.

* **Case-2** when **Alpha= 1**: 
Here the Regularization strength is decent but still the Loss function would get higher importance and hence the model is still trying to fit even the outliers. If we observe carefully, both in Case-1 and in Case-2, as the number of outliers are increasing, the hyperplane is trying to get as close as possible to the outliers and fit it.

* **Case-3** when **Alpha=100**:  
In this case the Regularization Strength is high. And therefore as the number of outliers are increasing, the hyperplane is not trying to reach the outliers and fit it. This is because the influence of the Loss term is reduced if the Regulularization strength is high and therefore the the model does not try to fit the outliers, there by getting rid of them.
